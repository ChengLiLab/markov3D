# Process ChIP-seq data

```python
import os
import sys
sys.path.append('/home/wangyn/lustre/pipeline/')
```

```python
import ngs.pipeline_chipseq
```

## GM12878 50kb

```python
# prefix format: cell_mark_resolution_chrom
resolution = 50e3
cell = 'GM12878'
```

```python
root = '/home/wangyn/lustre/ENCODE/ChIP_seq/GM12878/'
marks = os.listdir(root)
print(marks)
```

```python
for mark in marks:
    print('Processing {}...'.format(mark))
    bed_path = os.path.join(root, mark, 'merge.bed')
    prefix = '{}_{}_50kb'.format(cell.lower(), mark.lower())
    ngs.pipeline_chipseq.pipeline_chipseq(bed_path, resolution, prefix)
```
